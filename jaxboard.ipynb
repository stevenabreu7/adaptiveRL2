```python
import jax.numpy as jnp
import matplotlib.pyplot as plt
import glob
```

```python
log_folder = 'logs/InvertedPendulum-v4/0713_0609'
fig, axs = plt.subplots(3, 2, figsize=(12, 6), dpi=300)
fig.suptitle(log_folder)
axs = axs.flatten()
fnames = sorted(glob.glob(f'{log_folder}/*.npy'))
for idx, fname in enumerate(fnames):
    axs[idx].set_title(fname.split('/')[-1].split('.')[0])
    axs[idx].plot(jnp.load(fname))
fig.tight_layout();
```
